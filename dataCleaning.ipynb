```python
import pandas as pd
```

```python
df = pd.read_csv("info.csv")
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]
df
```

```python
# df = df.drop(["html"], axis=1)
df["title"] = df["title"].replace('<h1 class="sart-title">', "")
df
```

```python
df["title"] = df["title"].str.replace('<h1 class="sart-title">', "")
df["title"] = df["title"].str.replace('</h1>', "")
df["title"] = df["title"].str.replace('<h1 class="plus-title">', "")
df["title"] = df["title"].str.replace('\n\t\t\t', "")
df = df.drop_duplicates(subset=['title'], keep="first")
df
```

```python
df.info()
```

```python
df = df.dropna()
df = df.drop_duplicates()
df.info()
```

```python
#find prevalence of authors
international_authors = ['BBC News Mundo','BBC Mundo','EFE','Redacción AFP','Forbes','REDACCIÓN TECNO','Redacción C-Studio','Redacción']
df_national_authors = df.loc[~df['author'].isin(international_authors)]
df_national_authors["author"].value_counts().nlargest(10)
```

```python
## Most prevalent words
from collections import Counter
nonoWords = ['el', 'la', 'los', 'las', 'un', 'una', 'unos', 'unas', 'a', 'ante', 'bajo', 'cabe', 'con', 'contra', 'de', 'desde', 'en', 'entre', 'hacia', 'hasta', 'mediante', 'para', 'por', 'según', 'sin', 'so', 'sobre', 'tras', 'versus', 'vía', 'y', 'que', 'del', 'su', 'es', 'más', 'cómo', 'son', 'qué', 'que', 'se', 'no', '<h1', 'class="plus-title">', 'al', 'sus', 'ser', 'está', 'esta', 'te', 'mi']
common = Counter(" ".join(df["title"]).split()).most_common()
commonGood = []

temp = len(common)
for i in range(temp):
    word = common[i][0].lower()
    count = common[i][1]
    # print(common[i][0])
    if word not in nonoWords:
        if count > 10:
            commonGood.append(common[i])
    #     common.remove(word)


print(commonGood)
print(len(commonGood))
```
