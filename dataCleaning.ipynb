```python
## Most prevalent words
from collections import Counter
nonoWords = ['el', 'la', 'los', 'las', 'un', 'una', 'unos', 'unas', 'a', 'ante', 'bajo', 'cabe', 'con', 'contra', 'de', 'desde', 'en', 'entre', 'hacia', 'hasta', 'mediante', 'para', 'por', 'según', 'sin', 'so', 'sobre', 'tras', 'versus', 'vía', 'y', 'que', 'del', 'su', 'es', 'más', 'cómo', 'son', 'qué', 'que', 'se', 'no', '<h1', 'class="plus-title">', 'al', 'sus', 'ser', 'está', 'esta', 'te', 'mi']
common = Counter(" ".join(df["title"]).split()).most_common()
commonGood = []

temp = len(common)
for i in range(temp):
    word = common[i][0].lower()
    count = common[i][1]
    # print(common[i][0])
    if word not in nonoWords:
        if count > 10:
            commonGood.append(common[i])
    #     common.remove(word)


print(commonGood)
print(len(commonGood))
```

```python
# https://www.cdc.gov/museum/timeline/covid19.html
covidtimeline = (df['publishing_date'] > '2019-12-12') & (df['publishing_date'] <= '2022-06-08')
covid_time_articles = df.loc[covidtimeline]
print("percentage of articles produced during pandemic vs any other time: %"+str(len(covid_time_articles)/len(df)*100))
```

```python
keywords = ['covid','coronavirus','pandemi','vacuna','distancia']
covid_articles = pd.DataFrame(columns=df.columns)
for kw in keywords: 
    covid_articles = pd.concat([covid_articles, covid_time_articles[covid_time_articles['title'].str.contains(kw, case=False)]],axis=0)
covid_articles = covid_articles.drop_duplicates()
print("percentage of articles produced during pandemic about covid or the pandemic itself, in relation to articles produced in that time range: %"+str(len(covid_articles)/len(covid_time_articles)*100))
covid_articles.sort_values(by="publishing_date",ascending=True)
```
