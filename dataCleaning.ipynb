```python
from collections import Counter
nonoWords = ['el', 'la', 'los', 'las', 'un', 'una', 'unos', 'unas', 'a', 'ante', 'bajo', 'cabe', 'con', 'contra', 'de', 'desde', 'en', 'entre', 'hacia', 'hasta', 'mediante', 'para', 'por', 'según', 'sin', 'so', 'sobre', 'tras', 'versus', 'vía', 'y', 'que', 'del', 'su', 'es', 'más', 'cómo', 'son', 'qué', 'que', 'se', 'no', '<h1', 'class="plus-title">', 'al', 'sus', 'ser', 'está', 'esta', 'te', 'mi']
common = Counter(" ".join(df["title"]).split()).most_common()
commonGood = []

temp = len(common)
for i in range(temp):
    word = common[i][0].lower()
    count = common[i][1]
    # print(common[i][0])
    if word not in nonoWords:
        if count > 10:
            commonGood.append(common[i])
    #     common.remove(word)


print(commonGood)
print(len(commonGood))
```
